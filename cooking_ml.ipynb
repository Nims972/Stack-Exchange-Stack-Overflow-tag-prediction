```python
%%time
list_of_all_tags = []
for i in range(len(y_train)):
    row_tags = []
    start = 0
    for j in range(len(y_train[i])):
        if(y_train[i][j] == ' '):
            row_tags.append(y_train[i][start : j])
            list_of_all_tags.append(y_train[i][start : j])
            start = j+1
    row_tags.append(y_train[i][start : len(y_train[i])])
    list_of_all_tags.append(y_train[i][start : len(y_train[i])])
    y_train[i] = row_tags
```

```python
len(list_of_all_tags)
```

```python
list_of_all_unique_tags = set(list_of_all_tags)
```

```python
len(list_of_all_unique_tags)
```

```python
from collections import defaultdict
tags_count = defaultdict(int)
for i in list_of_all_tags:
    tags_count[i] += 1
```

```python
cc = 0
for tags, cnt in tags_count.items():
    if(cnt < 5):
        cc += 1
print(cc)
```

```python
tags_to_remove = [tags for tags, cnt in tags_count.items() if(cnt < 100)]
```

```python
tags_to_remove = set(tags_to_remove)
```

```python
from collections import defaultdict
type(tags_count)
for k, values in tags_count.items():
    print(values)
    print(k)
    
```

```python
%%time
y = []
for i in range(len(y_train)):
    j = 1
    new_tags = []
    new_tags.append(y_train[i][0])
    while(j < len(y_train[i])):
        if(not(y_train[i][j] in tags_to_remove)):
            new_tags.append(y_train[i][j])
        j += 1
    y.append(new_tags)
```

```python
final_tags = []
for i in range(len(y)):
    for j in range(len(y[i])):
        final_tags.append(y[i][j])
```

```python
len(final_tags)
final_tags = set(final_tags)
print(len(final_tags))
```

```python
%%time
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer(classes=list(final_tags))
y = mlb.fit_transform(y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
def tfidf_features(X_train, X_test):    
    tfidf_vectorizer = TfidfVectorizer(min_df=10, max_df=0.9, ngram_range=(1, 2),
                                       token_pattern='(\S+)')
    X_train = tfidf_vectorizer.fit_transform(X_train)
    X_test = tfidf_vectorizer.transform(X_test)
    
    return X_train, X_test, tfidf_vectorizer.vocabulary_
```
